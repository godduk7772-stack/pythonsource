````python
# 새로운 리스트 생성 : 리스트 컴프리핸션

print([i for i in range(1,7)])
print([i * i for i in range(5)])
print([i * 2 for i in[1,3,5,7,9]])
print([[i, i+1, i+2] for i in [1,3,5,7,9]])

# 1~10 사이의 숫자 중에서 홀수 리스트 생성
odd_list = []
for i in range(1,11): # 첫 번째 복붙
    if i % 2 == 1: # 두번 째 복붙
        odd_list.append(i)

odd_list

# 리스트 컴프리핸션으로 변환하기 # for 앞에 i와 같은 변수 입력,

odd_list2 = [i for i in range(1,11) if i % 2 == 1]
print(odd_list2)

````

````python
q_list= ["갑","을","병","정","경"]

#q_list 에서 정을 제외한 요소를 갖는 새로운 리스트 생성
# for


# print(q_list.pop(3))
# print(q_list)



# for a in q_list:
#        if a != "정":

#강사님 답안
new_list= []
for i in q_list:
       if i != '정':
              new_list.append(i)
print(new_list)

# 리스트 컴프리핸션

print([i for i in q_list if i != '정'])


````

````python
q1 = ['nice', 'study', 'python', 'anaconda','!']
#5글자 이상의 단어만 새로운 리스트로 생성
# len()
q2 = []

for i in q1:
    if len(i) >= 5:
        q2.append(i)

print(q2)
````
