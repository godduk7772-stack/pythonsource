````python
# 한 시간 임금은 10700원, 8시간 까지는 10700원으로 지급하고 8시간 초과 근무시에는 1.5배 지급
# 근무시간(hours), rate(시간당 임금), pay(일당)
hours,rate,pay =12,10700,0

#강사님 답안
if hours >8:
    pay = rate * 8 + ((hours-8) * rate * 1.5)
else:
    pay = rate * hours

print("근무시간:%d, 일당:%d" %(hours,pay))
print("근무시간 : {}, 일당 : {}".format(hours, pay)) #소수점이 나옴
print("근무시간 : {0}, 일당 : {1:.0f}".format(hours, pay)) #위의 경우 이처럼 하면 소수점이 안나옴
print(f"근무시간 : {hours}, 일당 : {pay}")
print(f"근무시간 : {hours}, 일당 : {pay}")
````
