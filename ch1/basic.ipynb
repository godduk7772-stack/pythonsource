#### 인덱싱과 슬라이싱
- 인덱스는 0번부터 시작

```python
a = "Life is too short, You need python"
print (a[33])
print (a[-2])
```

```python
# 슬라이싱 시 끝번호는 포함하지 않음

print(a[0:5])
print(a[0:2])
print(a[5:7])
print(a[12:17])
print(a[:])
print(a[19:-7])
print(a[19:])
print(a[:5])
```

```python
a="20260714Rainy"
#날짜와 날씨분리
print(a[:8])
print(a[8:])

#연월일 날씨분리
print(a[:4])
print(a[4:6])
print(a[6:8])
print(a[8:])
```
