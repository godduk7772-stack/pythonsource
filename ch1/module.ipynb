```python
# 쉼표로 구분된 파일 | < 이것도 포함
# 혹은 다양한 기호로 분리
import csv

with open("./data/sample1.csv","r")as f:
    reader = csv.reader(f) # txt 파일은 f.read 는 전부 읽는다.
    # print(reader)
    next(reader) # 헤더를 불러오고 싶지 않다면 좌측과 같이/ 헤더도 갖고 오고 싶다면 좌측 제거
    for c in reader:
        print(c)
```

```python
with open("./data/sample1.csv","r")as f:
    reader = csv.DictReader(f) # txt 파일은 f.read 는 전부 읽는다.
    # print(reader)
    next(reader) # 헤더를 종목별로 넣어주려면 좌측과 같이
    for c in reader:
        print(c)
```

```python
with open("./data/sample1.csv","r")as f:
    reader = csv.DictReader(f) # txt 파일은 f.read 는 전부 읽는다.
    # print(reader)
    next(reader) # 헤더를 종목별로 넣어주려면 좌측과 같이
    for c in reader:
        for k,v in c.items():
            print(k,v)
```

```python
import json

with open("./data/myinfo.json","r",encoding="utf-8") as f:
    data = json.load(f)
    print(data)
    print(type(data))
```

```python
dict1 = {'name':'성춘향', 'birth':'0927', 'age':24}

with open("./data/myinfo.json","w",encoding="utf-8") as f:
    json.dump(dict1, f,ensure_ascii=False)
```

```python
# loads() : json 문자열 => 파이썬 자료형(dict)으로 변환
# dumps()  파이썬 자료형(dict)으로 변환 => json 문자열 

str1 = """
{
    "name":"성춘향",
    "birth":"0727",
    "age":24
}
"""

data = json.loads(str1)
print(data)
print(type(data))

print(data["name"])
print(data.get("name"))
```

```python
dict1 = {
    "name":"성춘향",
    "birth":"0727",
    "age":20    
}

data=json.dumps(dict1)
print(data)
print(type(data))
```
