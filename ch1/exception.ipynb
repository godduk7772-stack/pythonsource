```python
#예외 클래스 생성 가능

class MyError(Exception):
    def __str__(self):
        return "허용되지 않는 별명입니다"
    

def say_nick(nick):
    if nick =="바보":
        raise MyError()
    print(nick)

try:
    say_nick("천사")
    say_nick("바보")
except: #except MyError as e: / print(e)
    print("오류해결")
```

```python
# 문제 1. 랜덤 비밀번호 생성기

# 사용자로부터 생성할 비밀번호의 길이를 입력받아 랜덤 비밀번호를 생성하는 프로그램 작성
#choice
chars="ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstvwxyz~~~~~~~01023456789!@#$%^&*"
# 비밀번호 길이는 4 이상만 허용
# 숫자가 아닌 값을 입력하면 예외 처리한다
import random

while True:
    try:
        pwd_num = int(input("비밀번호 길이 : "))

        if pwd_num < 4:
            print("비밀번호는 4자리 이상이어야 합니다.")
            continue
        break
    except ValueError:
        print("숫자만 입력하세요.")

password = []
for i in range(pwd_num):
    password.append(random.choice(chars))


password = [random.choice(chars) for i in range(pwd_num)]

# print(f"랜덤 비밀번호 : {password}")
#리스트 => 문자열로 변경

print(f"랜덤 비밀번호:{"".join(password)}")

# a=(int(input("원하는 비밀번호 길이 만큼 만들어드립니다.")))

# from random import random, sample

# if 4 <= a :
#     print(sample(chars, k=a))
# else:
#     print("4이상 입력해주세요.")
```

```python
# 나이 계산 프로그램
# 생년월일을 입력받아 현재 나이를 계싼하는 프로그램 작성
# 입력형식 2001-08-25
# input() => ' 2001-08-25' => 날짜 형식
# from datetime import datetime

# a=input("태어난 년도를 입력해 주세요.")
# print(a)
# print(type(a))



# # b=int(a)
# # print(a)

# # 문자 => 날짜
# birth = datetime.strptime(a,"%Y-%m-%d")
# print(type(birth))

# a=birth
# print(a)


#================강 사 님 답 안 ================================#
import datetime
# 오늘 날짜
now = datetime.datetime.now()
try:
    #생년월일 입력
    birth_str=input("생년월일 입력(YYYY-MM-DD)")
    # 2001-05-15
    birth = datetime.datetime.strptime(birth_str,"%Y-%m-%d")

    # 2026 - 2001
    # now 년도만 추출

    age = now.year - birth.year
    if (now.month, now.day) < (birth.month, birth.day):
        age -= 1

    print("나이",age)
except:
    print("날짜 형식이 안맞음")
```
