```python
# 문자열을 입력받아 다음 정보를 출력
# 출력내용 : 입력한 문자, 각 문자의 유니코드 값, 가장 큰 문자, 가장 작은 문자
a=input()

for i in a:
    if i == int:
        print(chr(i))

    elif i == str:
        print(ord(i))


# ord,chr,max,min,list
# 문자열 입력: Python
# P->80
# y->121

# 가장 큰 문자 : y
# 가장 작은 문자 : p



word = input("영어 문자열 입력:")
print(word)


for ch in word:
    print(f"{ch} -> {ord(ch)}")

print(f"가장 큰 문자: {max(word)}")
print(f"가장 작은 문자 :{min(word)}")
```

```python
# 사용자로부터 숫자 10개 입력받기

numbers = int(input)
for i in range(10):
    num = int(input(f"숫자 입력 :"))
    numbers.append(num)

print(f"원본 리스트 {numbers}")
unique_numbers = list(set(numbers))
sorted_numbers=sorted(unique_numbers)
print(f"중복 제거{set(unique_numbers)}")

print(f"오름차순 정렬{sorted_numbers}")
print(f"리스트 =튜플{tuple(sorted_numbers)}")
print(f"중복 제거 후 집합 크기{len(unique_numbers)}")


# 출력
# 입력리스트, 중복제거, 오름차순, 튜플변환, 집합 크기
# 3 5 2 5 7 ~

# 원본 [3,5,2,5~~]
# 중복제거 {1,2,3,4} set()
# 정렬 sorted()
# 튜플 tuple()
# 집합 크기 : len()
```
