#### file 읽고 쓰기 / 

|파일 열기 모드| 설명|
|--------------|-----|
|`r`| **읽기(Read) 모드** : 파일의 내용을 읽기만 할 때 사용,파일이 존재하지 않는 경우 오류 발생. |
|`w`| **쓰기(Write) 모드** : 파일의 내용을 새로 쓸 때 사용,기존파일이 존재하는 경우 내용 삭제후 새로 작성, 없는 경우 새로 생성 |
|`a`| **추가(Apppend) 모드** : 기존 파일의 마지막에 내용을 추가할 때 사용, 파일이 없는 경우 새로 생성 |
