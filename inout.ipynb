```python
# 1. 계산기 함수 먼저 정의
def calculator(n1, n2, operator):
    if operator == "+":
        return n1 + n2
    elif operator == "-":
        return n1 - n2
    elif operator == "*":
        return n1 * n2
    elif operator == "/":
        return n1 / n2
    else:
        return "지원하지 않는 연산자입니다."


# 2. 입력받기 (예: 15 10 +)
values = input("두 숫자와 연산자 기호 입력 (ex: 15 10 +): ")

# 3. 공백을 기준으로 분리 후 각각 변수에 담기
num1, num2, op = values.split()

# 4. 정수로 변환 후 함수 호출 및 출력
print(calculator(int(num1), int(num2), op))
    
#숫자 2개 입력받기, 연산자 기호(+,*...) 입력받기



# print(calculator(10,20,"+"))
# print(calculator(10,20,"-"))
# print(calculator(10,20,"*"))
# print(calculator(10,20,"/"))
# print(calculator(10,20,"//"))
```
