````python
# add
# add(2,3)
# add(2,3,4,5)
# add(2,3,4,5,6,7)

# 가변 매개변수(정해지지 않은 수를 더해주거나 처리할 때 쓰는 말)
# 입력값이 몇 개가 될 지 모르는 경우
# *
# def 함수명(*변수명):
#     해야할 일

def add(*args):
    result = 0
    for i in args:
        result = result + i
    print(result)
````

````python
# 키워드 매개변수 (key = value)
# ** 변수명

# key와 value(매개변수) 형태로 오면 딕셔너리 형태다.
def print_kwargs(**kwargs):
    print(kwargs)

    #key 값만 출력
    for k in kwargs.keys():
        print(k)
        
        #value 값만 출력
    for V in kwargs.values():
        print(V)
    
    #key, value 같이 출력
    for k,v in kwargs.items():
        print(k,v)
````

````python
print_kwargs(name="kim",age=15)
print_kwargs(name="park",addr="서울")
print_kwargs(name="hong",addr="경기", tel="010-1234-5678")
````

````python
# 전역변수 a 값 변경하기
# 1) 리턴 이용(추천방법)

a=1
def vartest(a):
    #매개변수 a :지역변수
    a=a+1
    return a

a=vartest(a)
print(a)
````

````python
# 2) global 키워드 사용
# 함수 안에서 전역변수 접근 안됨

a=1
def vartest():
    #a값을 불러와서 사용하려면 아래와 같이. global 을 써주고 해당 변수 a를 입력해준다.
    global a
    a=a+1

vartest()
print(a)
````

````python
# 사칙연산 함수
# 결과 리턴
def calculator(num1,num2,op):
    # if~elif
    if op == "+":
        return num1 + num2
    elif op == "-":
        return num1 - num2
    elif op == "*":
        return num1 * num2
    elif op == "/":
        return num1 / num2
    elif op == "/":
        return num1 // num2

print(calculator(10,20,"+"))
print(calculator(10,20,"-"))
print(calculator(10,20,"*"))
print(calculator(10,20,"/"))
print(calculator(10,20,"//"))
````
